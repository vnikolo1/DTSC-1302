```python
import pandas as pd
import statsmodels.api as sm

dataset = pd.read_csv("DataCleaning_2.csv")
df = pd.DataFrame(dataset)
```

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import statsmodels.formula.api as smf

response = 'GradRate'
predictors = df.drop(columns=[response])

correlation = predictors.corr()
print(correlation)

plt.figure(figsize = (5,3))
sns.heatmap(df.corr(), annot= True, cmap = "coolwarm", vmin=-1, vmax=1)
plt.title("Correlation of PHDs, Term, SF Ratio and Expenditure with Graduation Rates - Public Unis")
plt.show()
```

```python
model_f = smf.ols(formula = "GradRate ~ PHDs + Terminal", data=df)

result = model_f.fit()

result.summary()
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import statsmodels.api as sm
import matplotlib.pyplot as plt
import statsmodels.formula.api as smf
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
dataset = pd.read_csv("DataCleaning_2.csv")
df = pd.DataFrame(dataset)
```

```python
percentages=['PHDs', 'Terminal', 'GradRate']
raw_columns = ['S.F. Ratio', 'Expend']

scaler = StandardScaler()

df[raw_columns] = scaler.fit_transform(df[raw_columns])

print(df)
```

```python
correlation_matrix = df.corr()

print("Correlation Matrix of PHDs, Terminal Degrees, SF Ratio and Expenditure with Graduation Rates:\n", correlation_matrix)

plt.figure(figsize=(5,3))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt = '.2f', linewidths = 1)
plt.title("Correlation Matrix of PHDs, Terminal Degrees, SF Ratio,\n and Expenditure with Graduation Rates - Public Unis")
plt.show()
```

```python
model_f = smf.ols(formula = "GradRate ~ PHDs + Terminal", data=df)

result = model_f.fit()

result.summary()
```
